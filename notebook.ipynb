# Scrapping financial data from screener.in

```python
from urllib.request import urlopen
import web_scrapping as ws
import pandas as pd
# import numpy as np
```

```python
link = "https://www.screener.in/company/IOC/consolidated/"

page = urlopen(link)

html_bytes = page.read()
html = html_bytes.decode("utf-8")
```

## Quaterly Results

```python
headings_value, part_html, rest = ws.getting_headers(html, "Quarterly Results", "Profit & Loss")

## Subsetting the dataset for quaterly results

# Quaterly sales
sales, temp = ws.func(part_html, "Sales&nbsp;", "td", headings_value)

# Quaterly expenses
expenses, temp = ws.func(temp, "Expenses&nbsp;", "td", headings_value)

# Quaterly operating_profit
operating_profit, temp = ws.func(temp, "Operating Profit", "td", headings_value)

# Quaterly OPM_per
opm_per, temp = ws.func(temp, "OPM %", "td", headings_value)

# Quaterly other_income
other_income, temp = ws.func(temp, "Other Income", "td", headings_value)

# Quaterly interest
interest, temp = ws.func(temp, "Interest", "td", headings_value)

# Quaterly depreciation
depreciation, temp = ws.func(temp, "Depreciation", "td", headings_value)

# Quaterly profit before tax
profit_before_tax, temp = ws.func(temp, "Profit before tax", "td", headings_value)

# Quaterly tax%
tax_per, temp = ws.func(temp, "Tax %", "td", headings_value)

# Quaterly net profit
net_profit, temp = ws.func(temp, "Net Profit", "td", headings_value)

# Quaterly EPS
eps, temp = ws.func(temp, "EPS in Rs", "td", headings_value)
```

```python
# Creating final dataframe
qr_index = ['sales', 'expenses', 'operating_profit', 'opm_per', 'other_income', 'interest', 
            'depreciation', 'profit_before_tax', 'tax_per', 'net_profit', 'eps']
l2 = [sales, expenses, operating_profit, opm_per, other_income, interest, depreciation,
      profit_before_tax, tax_per, net_profit, eps]

for i in range(len(l2)):
    l2[i] = ws.string_to_float(l2[i])


quater_results = pd.DataFrame({'index':qr_index },columns = headings_value)

for i in range(len(l2)):
    quater_results.loc[i] = l2[i]

quater_results['parameter'] = qr_index
quater_results = quater_results.set_index('parameter')
```
