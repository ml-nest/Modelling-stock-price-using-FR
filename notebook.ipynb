# Scrapping financial data from screener.in

```python
from urllib.request import urlopen
import web_scrapping as ws
import pandas as pd
# import numpy as np
```

```python
link = "https://www.screener.in/company/IOC/consolidated/"

page = urlopen(link)

html_bytes = page.read()
html = html_bytes.decode("utf-8")
```

## Quaterly Results

```python
headings_value, temp, rest = ws.getting_headers(html, "Quarterly Results", "Profit & Loss")

qr_index = ['Sales', 'Expenses', 'Operating Profit', 'OPM %', 'Other Income', 'Interest', 
            'Depreciation', 'Profit before tax', 'Tax %', 'Net Profit', 'EPS in Rs']

quarter_results = pd.DataFrame(columns = headings_value)

for i in qr_index:
    values, temp = ws.func(temp, i, "td", headings_value)
    quarter_results.loc[i] = ws.string_to_float(values)
```
