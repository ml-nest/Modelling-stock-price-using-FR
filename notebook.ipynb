### Scrapping financial data from screener.in

```python
from urllib.request import urlopen
```

```python
link = "https://www.screener.in/company/IOC/consolidated/"

page = urlopen(link)

html_bytes = page.read()
html = html_bytes.decode("utf-8")
```

```python
def results(temp, start_text, end_text):    
    val1 = temp[temp.find(start_text)+len(start_text):temp.find(end_text)]
    temp = temp[temp.find(end_text)+len(end_text):]
    return val1, temp

def func(html, find_text, breaker):
    headings = html.find(find_text)

    temp = html[headings:]
    temp = temp[temp.find("/"+breaker+">")+len("</"+breaker+">"):]

    start_text, end_text = "<"+breaker+">", "</"+breaker+">"

    headings = []
    for i in range(10):
        q, temp = results(temp, start_text, end_text)
        headings.append(q)
    return headings, temp
```

```python
## Quaterly Results


# Quaterly headers
headers, temp = func(html, "Quarterly Results", "th")

# Quaterly sales
sales, temp = func(temp, "Sales&nbsp;", "td")

# Quaterly expenses
expenses, temp = func(temp, "Expenses&nbsp;", "td")

# Quaterly operating_profit
operating_profit, temp = func(temp, "Operating Profit", "td")

# Quaterly OPM_per
opm_per, temp = func(temp, "OPM %", "td")

# Quaterly other_income
other_income, temp = func(temp, "Other Income", "td")

# Quaterly interest
interest, temp = func(temp, "Interest", "td")

# Quaterly depreciation
depreciation, temp = func(temp, "Depreciation", "td")

# Quaterly profit before tax
profit_before_tax, temp = func(temp, "Profit before tax", "td")

# Quaterly tax%
tax_per, temp = func(temp, "Tax %", "td")

# Quaterly net profit
net_profit, temp = func(temp, "Net Profit", "td")

# Quaterly net profit
eps, temp = func(temp, "EPS in Rs", "td")
```
